## PTA

```python
link="https://pta.trunojoyo.ac.id/c_search/byprod/10/"
```

```python
temppta=[]
# data=0
for i in range(152, 153):
  print('Page -', i)
  webpage = requests.get(link+str(i))
  soup = BeautifulSoup(webpage.text, "html.parser")
  items =soup.find_all('li', {'data-cat':'#luxury'})

  for it in items:
    # data+=1
    linkjour = it.find('a', 'gray button')['href']
    # print(linkjour)

    webpage2 = requests.get(linkjour)
    soup2 = BeautifulSoup(webpage2.text, "html.parser")
    nim = ''.join(linkjour).replace('https://pta.trunojoyo.ac.id/welcome/detail/', '')
    judul = ''.join(it.a.text.replace('\r', '').split('\n'))
    abstrak = soup2.p.text
    nama = it.span.text.replace('Penulis : ', '')

    temppta.append([nim, nama, judul, abstrak])

print('Totaldata', len(temppta))
```

```python
headCsvPta=['NIM', 'Nama', 'Judul', 'Abstrak']
dfPta = pd.DataFrame(temppta, columns=headCsvPta)
```

## Twitter

```python
import snscrape.modules.twitter as sntwitter

query = "ridwan kamil presiden since:2022-01-01 until:2023-02-20"
tweets = []
limit = 2000


for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    
    # print(vars(tweet))
    # break
    if len(tweets) == limit:
        break
    else:
        # td=tweet.date
        # td=td.strftime("%Y-%m-%d %H:%M:%S")
        tweets.append([ tweet.username, tweet.content, tweet.user.location, tweet.date.strftime("%Y-%m-%d %H:%M:%S")])
```

```python
dfTwit = pd.DataFrame(tweets, columns=[ 'Username', 'Twitts','Lokasi', 'Datetime'])
```

## Berita

```python
linkberita='https://news.kompas.com/search/all/'
```

```python
berita=[]

for i in range(1,9):
  print('Page -', i)
  req = requests.get(linkberita+str(i))
  soup = BeautifulSoup(req.text, 'html.parser')
  items = soup.findAll('div', 'article__list clearfix')

  for it in items:
    link = it.find('a','article__link')['href']


    req2 = requests.get(link)
    soup2 = BeautifulSoup(req2.text, 'html.parser')
    # items2 = soup2.findAll('div', 'article__list clearfix')
    judul = soup2.find('h1', 'read__title').text
    isi = ''.join(soup2.find('div', 'read__content').text.split('\n')).replace('JAKARTA, KOMPAS.com - ', '').replace('Dapatkan update berita pilihan dan breaking news setiap hari dari Kompas.com. Mari bergabung di Grup Telegram "Kompas.com News Update", caranya klik link https://t.me/kompascomupdate, kemudian join. Anda harus install aplikasi Telegram terlebih dulu di ponsel.','')
    # isifix=''.join(isi[:])
    berita.append([judul,isi])
    # print(judul)
    # print(isi)
```

```python
dfBerita = pd.DataFrame(berita, columns=[ 'Judul', 'Isi'])
```
