## Pipeline

### Ambil data 

1. install database (MySql, Sql Server, Postgre, Postgre elephant)
2. masukkan data iris ke tiap database, dengan tambahan id untuk identitas
3. ambil data dari database:
*   PostgreSql Local(SepalLength)
*   PostgreSql Elephant(SepalWidth)
*   MySql Local (PetalLength)
*   Sql Server(PetalWidth, Class)
4. pengambilan data menggunakan library psycopg2, mysql.connector, pyodbc
5. tiap server mengambil kolom sesuai rule
6. kolom yang diambil dari tiap server di convert ke csv dengan modul pandas

### Pipeline

1. Masuk ke https://www.orchest.io/ , Login dan buat project baru
2. upload file csv yang didapat ke folder dataset pipeline
3. buat new step untuk GetData, Splitting Data, Modelling dan Hasil
4. Code GetData untuk mengambil data tiap kolom dan di gabung menjadi satu
5. SplitData untuk mambagi data dari GetData menjadi X_train, X_test, y_train, y_test
6. Data train dan tes dikirim ke step modelling untuk diolah dan mendapatkan hasil akurasi
7. Hasil akurasi terdapat pada step hasil
